## Data Analysis of Plane Crashes

```python
import pandas as pd
```

```python
d = pd.read_csv('plane.csv')
```

```python
# 1. Sort the plane crashes by quantity of total fatalities based on time and date.

print("Plane Crashes Based on Time and Date:")
gd = d.groupby(['Date', 'Time'])['Total fatalities'].sum()
sd = gd.sort_values(ascending=False)
sd.head(20)
```

```python
# 2. Sort the plane crashes by quantity of survivors (Yes) based on region.

print("Plane Crashes Sorted By Survivors Who Made It Alive Based on Region")
yes = d[d['Survivors'] == 'Yes']
gd = yes.groupby('Region')['Survivors'].count()
sd = gd.sort_values(ascending=False)
sd
```

```python
# 3. Sort the plane crashes by quantity of survivors (No) based on region.

print("These People Did Not Make It Alive in a Plane Crash")
no = d[d['Survivors'] == 'No']
gd = no.groupby('Region')['Survivors'].count()
sd = gd.sort_values(ascending=False)
sd
```

```python
# 4. Sort the plane crashes by quantity of crash causes (Technical Failure, Human Factor, Weather, Unknown, Terrorism, Other Causes) by country.

print("Plane Crashes Sorted by Crash Causes")
cause = ['Technical failure', 'Human factor', 'Weather', 'Unknown', 'Terrorism act', 'Hijacking', 'Sabotage', 'Other causes']
data = d[d['Crash cause'].isin(cause)]
gd = data.groupby(['Country', 'Crash cause'])['Crash cause'].count()
ud = gd.unstack(fill_value=0)
ud['Total'] = ud.sum(axis=1)
sd = ud.sort_values('Total', ascending=False)
sd
```

```python
# 5. Sort the plane crashes by quantity of flight types based on country.

print("Plane Crashes Sorted by Flight Types Based on Country")
flighttype = ['Military', 'Training', 'Test', 'Delivery', 'Ferry', 'Cargo', 'Ambulance', 'Cinematography', 'Supply', 'Bombing', 'Demonstration', 'Refuelling', 'Positioning', 'Charter/Taxi', 'Survey / Patrol / Reconnaissance', 'Private', 'Scheduled Revenue Flight', 'Executive/Corporate/Business', 'Postal (mail)', 'Government', 'Aerial photography']
data = d[d['Flight type'].isin(flighttype)]
gd = data.groupby(['Country', 'Flight type'])['Flight type'].count()
ud = gd.unstack(fill_value=0)
ud['Total'] = ud.sum(axis=1)
sd = ud.sort_values('Total', ascending=False)
sd
```

```python
# 6. Sort the plane crashes by quantity of flight phases based on country.

print("Plane Crashes Sorted by Flight Phases Based on Country")
flightphase = ['Takeoff (climb)', 'Flight', 'Landing (descent or approach)', 'NA']
data = d[d['Flight phase'].isin(flightphase)]
gd = data.groupby(['Country', 'Flight phase'])['Flight phase'].count()
ud = gd.unstack(fill_value=0)
ud['Total'] = ud.sum(axis=1)
sd = ud.sort_values('Total', ascending=False)
sd
```

```python
# 7. Sort the plane crashes by the highest crew on board and highest pax on board based on country.

print("Plane Crashes Sorted by Highest Crew and Pax on Board Based on Country")
gd = d.groupby('Country')[['Crew on board', 'Pax on board']].max()
sdc = gd.sort_values('Crew on board', ascending=False)
sdp = gd.sort_values('Pax on board', ascending=False)
sdc.head(10)
sdp.head(10)
```

```python
# 8. Sort the plane crashes by the lowest crew on board and lowest pax on board based on country.

print("Plane Crashes Sorted by Lowest Crew and Pax on Board Based on Country")
gd = d.groupby('Country')[['Crew on board', 'Pax on board']].max()
sdc = gd.sort_values('Crew on board', ascending=True)
sdp = gd.sort_values('Pax on board', ascending=True)
sdc.head(10)
sdp.head(10)
```

```python
# 9. Sort the plane crashes by quantity of total fatalities based on aircraft, operator, and registration.

print("Plane Crashes Sorted by Total Fatalities Based on Aircraft, Operator, and Registration")
gd = d.groupby(['Aircraft', 'Operator', 'Registration'])['Total fatalities'].sum()
sd = gd.sort_values(ascending=False)
sd.head(10)
```

```python
#10. Sort the plane crashes by quantity of total fatalities based on crash site and crash location.

print("Plane Crashes Sorted by Total Fatalities Based on Crash Site and Crash Location")
gd = d.groupby(['Crash site', 'Crash location'])['Total fatalities'].sum()
sd = gd.sort_values(ascending=False)
sd.head(10)
```

```python
import hvplot.pandas
import holoviews as hv
import bokeh
hv.extension('bokeh')
```

```python
# 1. Show the plane crashes by quantity of total fatalities based on date using horizontal bar.

gd = d.groupby(['Date', 'Time'])['Total fatalities'].sum()
sd = gd.sort_values(ascending=False).head(25)
sd.hvplot.bar(x='Date', y='Total fatalities', color='black', title='Top 25 Total Fatalities Based on Date')
```

```python
# 2. Show the plane crashes by quantity of survivors (Yes) based on region using vertical bar.

yes = d[d['Survivors'] == 'Yes']
gd = yes.groupby('Region')['Survivors'].count()
sd = gd.sort_values(ascending=False)
sd.hvplot.bar(x='Region', y='Survivors', ylabel='Survivors', color='black', height=500, width=800, label="There Are Survivors")
```

```python
# 3. Show the plane crashes by quantity of survivors (No) based on region using vertical bar.

no = d[d['Survivors'] == 'No']
gd = no.groupby('Region')['Survivors'].count()
sd = gd.sort_values(ascending=False)
sd.hvplot.line(x='Region', y='Survivors', ylabel='Corpses', color='black', height=500, width=800, label="There Are No Survivors")
```

```python
# 4. Show the plane crashes by quantity of crash causes based country using histogram.

cause = ['Technical failure', 'Human factor', 'Weather', 'Unknown', 'Terrorism act, Hijacking, Sabotage', 'Other causes']
data = d[d['Crash cause'].isin(cause)]
gd = data.groupby(['Country', 'Crash cause'])['Crash cause'].count()
ud = gd.unstack(fill_value=0)
ud['Total'] = ud.sum(axis=1)
sd = ud.sort_values('Total', ascending=False)

sd.hvplot.bar(
    x='Country', 
    y=['Technical failure', 'Human factor', 'Weather', 'Unknown', 'Terrorism act, Hijacking, Sabotage', 'Other causes'], 
    stacked=True, 
    height=500, 
    width=800,
    ylabel='Number of crashes'
)
```

```python
# 5. Show the plane crashes by quantity of flight types based on country using scatter plot.

flighttype = ['Military', 'Training', 'Test', 'Delivery', 'Ferry', 'Cargo', 'Ambulance', 'Cinematography', 'Supply', 'Bombing', 'Demonstration', 'Refuelling', 'Positioning', 'Survey / Patrol / Reconnaissance', 'Private', 'Scheduled Revenue Flight', 'Executive/Corporate/Business', 'Postal (mail)', 'Government', 'Aerial photography']
data = d[d['Flight type'].isin(flighttype)]
gd = data.groupby(['Country', 'Flight type'])['Flight type'].count()
ud = gd.unstack(fill_value=0)
ud['Total'] = ud.sum(axis=1)
sd = ud.sort_values('Total', ascending=False)

sd.hvplot.bar(
    x='Country', 
    y=['Military', 'Training', 'Test', 'Delivery', 'Ferry', 'Cargo', 'Ambulance', 'Cinematography', 'Supply', 'Bombing', 'Demonstration', 'Refuelling', 'Positioning', 'Survey / Patrol / Reconnaissance', 'Private', 'Scheduled Revenue Flight', 'Executive/Corporate/Business', 'Postal (mail)', 'Government', 'Aerial photography'],
    stacked=True, 
    height=500, 
    width=800,
    ylabel='Number of crashes'
)
```

```python
# 6. Show the plane crashes by quantity of flight phases based on country scatter plot.

flightphase = ['Takeoff (climb)', 'Flight', 'Landing (descent or approach)']
data = d[d['Flight phase'].isin(flightphase)]
gd = data.groupby(['Country', 'Flight phase'])['Flight phase'].count()
ud = gd.unstack(fill_value=0)
ud['Total'] = ud.sum(axis=1)
sd = ud.sort_values('Total', ascending=False)
sd = sd.reset_index()

sd.hvplot.scatter(
    x='Country',
    y=['Takeoff (climb)', 'Flight', 'Landing (descent or approach)'],
    color=['red', 'green', 'blue'],
    height=500,
    width=800,
    xlabel='Country',
    ylabel='Number of crashes',
    title='Flight Phase vs Number of Crashes',
    alpha=0.5,
    marker='o'
)
```

```python
# 7. Show the plane crashes by the highest crew on board and highest pax on board based on country using stacked bar.

gd = d.groupby('Country')[['Crew on board', 'Pax on board']].max()
sdc = gd.sort_values('Crew on board', ascending=False)
sdp = gd.sort_values('Pax on board', ascending=False)

sdc_top_10 = sdc.head(10)
sdc_top_10.hvplot(kind='bar', stacked=True, x='Country', y=['Crew on board', 'Pax on board'])
```

```python
# 8. Show the plane crashes by the lowest crew on board and lowest pax on board based on country using stacked bar.

gd = d.groupby('Country')[['Crew on board', 'Pax on board']].max()
sdc = gd.sort_values('Crew on board', ascending=True)
sdp = gd.sort_values('Pax on board', ascending=True)

sdc_top_10 = sdc.head(10)
sdc_top_10.hvplot(kind='bar', stacked=True, x='Country', y=['Crew on board', 'Pax on board'])
```

```python
# 9. Show the plane crashes by quantity of total fatalities based on aircraft using unstacked bar chart.

gd = d.groupby(['Aircraft', 'Operator', 'Registration'])['Total fatalities'].sum()
sd = gd.sort_values(ascending=False).head(10)
sd.hvplot(kind='bar', stacked=False, xlabel='Aircraft', color='Gray', title='Top 10 Fatal Accidents by Aircraft')
```

```python
# 10. Show the plane crashes by quantity of total fatalities based on crash site and crash location using heatmap plot.

gd = d.groupby(['Crash site', 'Crash location'])['Total fatalities'].sum()
sd = gd.sort_values(ascending=False).head(10)

df = pd.DataFrame(sd.reset_index())
heatmap = df.hvplot.heatmap(x='Crash location', y='Crash site', C='Total fatalities', cmap='Reds', height=500, title='Top 10 crash locations by total fatalities')

heatmap
```
